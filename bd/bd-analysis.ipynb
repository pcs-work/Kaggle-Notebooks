```python
import os
import cv2
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def show(
    image_1: np.ndarray,
    image_2: np.ndarray,
    cmap: str="gnuplot2",
) -> None:
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(image_1, cmap=cmap)
    plt.axis("off")
    plt.subplot(1, 2, 2)
    plt.imshow(image_2, cmap=cmap)
    plt.axis("off")
    plt.show()
```

```python
r.seed(42)
for _ in range(5):
    FILENAME: str = str(r.randint(0, 1000)).zfill(4)    
    
    image = cv2.cvtColor(src=cv2.imread(f"/kaggle/input/birdies/images/{FILENAME}.jpg", cv2.IMREAD_COLOR), code=cv2.COLOR_BGR2RGB)

    with open(f"/kaggle/input/birdies/labels/{FILENAME}.txt", "r") as fp:
        data = fp.read()
    data = data.split(" ")

    cx = int(float(data[1]) * image.shape[1])
    cy = int(float(data[2]) * image.shape[0])
    w  = int(float(data[3]) * image.shape[1])
    h  = int(float(data[4]) * image.shape[0])

    x_min = cx - w//2
    y_min = cy - h//2
    x_max = cx + w//2
    y_max = cy + h//2

    disp_image = image.copy()
    cv2.rectangle(disp_image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 3)

    show(image, disp_image)
```

```python
r.seed(42)
SIZE: int = 224

for _ in range(5):
    FILENAME: str = str(r.randint(0, 1000)).zfill(4)    
    
    image = cv2.resize(src=cv2.cvtColor(src=cv2.imread(f"/kaggle/input/birdies/images/{FILENAME}.jpg", cv2.IMREAD_COLOR), code=cv2.COLOR_BGR2RGB), dsize=(SIZE, SIZE), interpolation=cv2.INTER_AREA)

    with open(f"/kaggle/input/birdies/labels/{FILENAME}.txt", "r") as fp:
        data = fp.read()
    data = data.split(" ")

    cx = int(float(data[1]) * SIZE)
    cy = int(float(data[2]) * SIZE)
    w  = int(float(data[3]) * SIZE)
    h  = int(float(data[4]) * SIZE)

    x_min = cx - w//2
    y_min = cy - h//2
    x_max = cx + w//2
    y_max = cy + h//2

    disp_image = image.copy()
    cv2.rectangle(disp_image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

    show(image, disp_image)
```
